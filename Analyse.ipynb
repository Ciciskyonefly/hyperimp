```python
def plot_marginal(x, avg, std, title=None, xname=None, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(x, avg)
    ax.fill_between(x, avg + std, avg - std, alpha=0.3)
    if xname is not None:
        ax.set_xlabel(xname)
    if title is not None:
        ax.set_title(title)
    ax.set_ylabel('error')
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

def plot_var_fractions(fracs, names):
    fig, ax = plt.subplots(figsize=(5,5))
    x = fracs + [1. - sum(fracs)]
    labels = names + ['higher order']
    explode = np.zeros_like(x)
    explode[-1] = 0.1
    _ = ax.pie(x, labels=labels, startangle=90, explode=explode)
```

```python
def auto_analyse_csv(filename):
    data = np.loadtxt(filename, delimiter=', ')
    X = data[:, :6]
    y = data[:, 6]
    r = analyse(X, y, names)
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 8))
    for i, k in enumerate(names):
        title = "%s = %.2f%%" % (k, r['marginals'][k] * 100)
        f = plot_marginal(*r['avg_stds'][k], title=title, ax=axes[i//3, i%3])
    return r
```

```python
def gen_reverse_config(ex, copykeys):
    conf = ex['config']
    rev_conf = {}
    for k in copykeys:
        set_by_dotted_path(rev_conf, k, get_by_dotted_path(conf, k, None))
    rev_conf['clip_gradient'] = int(conf['clip_gradient'])       # 4
    rev_conf['hidden_size'] = np.log10(conf['hidden_size'] / 2)  # 2
    rev_conf['input_noise_std'] = conf['input_noise_std']        # 3
    rev_conf['learning_rate'] = np.log10(conf['learning_rate'])  # 0
    rev_conf['momentum'] = np.log10(1 - conf['momentum'])        # 1
    rev_conf['use_nesterov'] = int(conf['use_nesterov'])         # 5
    
    rev_conf['lstm_configuration']['in_act_func'] = (conf['lstm_configuration'].get('in_act_func') != 'linear')
    rev_conf['lstm_configuration']['act_func'] = (conf['lstm_configuration'].get('act_func') != 'linear')
    rev_conf['result'] = ex['result']
    return rev_conf
```

```python
def query_data(db, query, keys, defaults=None):
    if defaults is None:
        defaults = {}
    query_done = copy(query)
    query_done['status'] = 'COMPLETED'
    exps = db.find(query_done)
    lines = []
    results = []
    for ex in exps:
        rcfg = gen_reverse_config(ex, keys)
        if not np.isfinite(rcfg['result']):
            continue
        lines.append([get_by_dotted_path(rcfg, k, defaults.get(k)) for k in keys])
        results.append(rcfg['result'])
    X = np.array(lines)
    y = np.array(results)
    return X, y
    

def auto_analyse_query(db, query, keys, defaults=None, n_trees=100):
    X, y = query_data(db, query, keys, defaults)
    
    print('%d Results obtained, analysing...' % y.shape[0])
    r = analyse(X, y, keys, n_trees=n_trees)
    plot_var_fractions([r['marginals'][k] for k in keys], keys)
    ncols = 3
    nrows = (len(keys) + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*5, nrows*5))
    for i, k in enumerate(keys):
        title = "%s = %.2f%%" % (k, r['marginals'][k] * 100)
        f = plot_marginal(*r['avg_stds'][k], title=title, ax=axes[i//ncols, i%ncols])
```
